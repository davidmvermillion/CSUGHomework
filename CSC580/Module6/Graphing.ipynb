```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
hist = pd.read_pickle("hist.pkl")
```

```python
fig, ax = plt.subplots()
plt.grid(which = 'major', axis = 'both', linestyle = ':', color = '#e9e9e9')
plt.plot(hist['epoch'], hist['accuracy'], color = '#D8973C', linestyle = (0, (5, 10)))
plt.plot(hist['epoch'], hist['val_accuracy'], color = '#273E47')
ax.set_xlabel('Epoch', fontsize = 15,
			  loc = 'left').set_color('#707070')
ax.set_ylabel('Accuracy', fontsize = 15, rotation = 'horizontal',
			  loc = 'bottom', labelpad = 75).set_color('#707070')
ax.set_title('Validation Accuracy Leveled\nAfter 7 Epochs', fontsize = 23, pad = 15).set_color('#171819')
for pos in ['right', 'top']:
	plt.gca().spines[pos].set_visible(False)
ax.spines['bottom'].set_color('#A0A0A0')
ax.spines['left'].set_color('#A0A0A0')
ax.tick_params(axis='y', which='both', right=False,
			   left=False, colors = '#686868')
ax.tick_params(axis='x', which='both', top=False,
			   bottom=False, colors = '#686868')

for line, name in zip(ax.lines, ['Training', 'Validation']):
    y = line.get_ydata()[-1]
    ax.annotate(name, xy=(1,y), xytext=(6,0), color = line.get_color(),
                xycoords = ax.get_yaxis_transform(), textcoords = "offset points",
                size = 14, va = "center")

plt.show();
```
